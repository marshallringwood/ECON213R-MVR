```python
#i will consider outliers on all variables--although there aren't many
q_df = cmc_df.quantile([.25, .75])
q_df.loc['iqr'] = q_df.loc[0.75] - q_df.loc[0.25]
q_df.loc['whisker_length'] = 1.5 * q_df.loc['iqr']
q_df.loc['max_whisker'] = q_df.loc['whisker_length'] + q_df.loc[0.75]
q_df.loc['min_whisker'] = q_df.loc[0.25] - q_df.loc['whisker_length']
q_df
```

```python
#i will reconsider potential outliers under "Number of children ever born" using Std dev. method
std_children = cmc_df['Number of children ever born'].std()
std_children
```

```python
max_outlier_children_value = cmc_df['Number of children ever born'].mean() + (std_children * 3)
max_outlier_children_value
```

```python
#from the chart, still dealing with outliers under "Number of children ever born"
max_outlier_children = [num for num in cmc_df['Number of children ever born'] if num > q_df['Number of children ever born'].loc['max_whisker']]
len(max_outlier_children)
```
